```python
import matplotlib.pyplot as plt
%matplotlib inline
```

**'%matplotlib inline' is useful for ploting something in jupyter notebook without the statement - 'plt.show()'**

```python
import numpy as np
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

## Scatter Plot

```python
plt.scatter(x,y,color='green')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('Test.png') # This will save the figure in the working dir as png format
```

## Line Plot

```python
y=x*x
```

```python
plt.plot(x,y,linewidth=2, markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2d Diagram')
```

## Subplot - plot

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'yo')
```

## sin - cos wave with plot

```python
x = np.arange(0, 5 * np.pi, 0.1) # Ranging 0 - 5*pi with 0.1 interval 
y_sin = np.sin(x) 
y_cos = np.cos(x)  

# First subplot 
plt.subplot(2, 1, 1)
plt.plot(x, y_sin,'g') 
plt.title('Sine')  
   
# Second subplot
plt.subplot(2, 1, 2) 
plt.plot(x, y_cos,'b') 
plt.title('Cosine')  
```

## Bar plot

```python
x = [2,8,10] 
y = [11,16,9]  

x2 = [3,9,11] 
y2 = [6,15,7] 
plt.bar(x, y) 
plt.bar(x2, y2, color = 'g') 
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')  
```

## Histograms

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a) 
plt.title("Histogram") 
```

## Box Plot

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.boxplot(data,vert=True,patch_artist=True); 
```

## Pie Chart

```python
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True)

plt.axis('equal')
```
