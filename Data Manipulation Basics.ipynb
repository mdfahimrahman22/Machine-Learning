# Reading CSV File

```python
df=pd.read_csv("uber drives.csv") # df assumed as dataframe 

print("Print starting and ending index with step count : ")
print(df.index)   # Print starting and ending index with step count
print("\nThe data types of the cells : ")
print(df.dtypes)  # The data types of the cells
print("\nThe data types of the cells : ")
print(df.shape)   # Print (rows,columns)
print("\nAll information about data pack : ")
print(df.info())   #All information about data pack
df.head(7) # 1st 7 rows //df.head() will print 1st 5 rows  
```

# Setting 1st  2columns into Date format

```python
df["START_DATE"]=pd.to_datetime(df["START_DATE"]) # Type casting "START_DATE" column
df["END_DATE"]=pd.to_datetime(df["END_DATE"])      
pd.to_datetime(df["START_DATE"],format='%d-%m-%Y %H:%M') # Correcting the format which is by default Year-Month-Day Hour:Min 
pd.to_datetime(df["END_DATE"],format="%d-%m-%Y %H:%M")
```

# Some Data analysis

```python
print("Frequencies of the row values from \"Stop\" column : ")
print(df["STOP"].value_counts()) # Prints the frequencies of the row values from "STOP" column
df.describe(include=[np.number]) # Describes only numeric values
df.describe(include=['number'])  # Describes only numeric values
df.describe(include=[np.object]) # Describes only objects
df.describe(exclude=[np.number]) # Describes all columns except numeric data columns
df.describe(include='all')  # Describe all columns
df.tail()   # Last 5 rows 
```

# Data Manipulation part-1

```python
#df=df.iloc[:-1,:] # Last row remove
df2=df.loc[[0,4],["START_DATE","MILES"]] # 0,4 index rows only including "START_DATE","MILES" columns 
df2.head()
 
df3=df.loc[df["MILES"]>10]  # rows where "MILES">10 //df3=df[df["MILES"]>10] do the same 
df3.head()

df.loc[df["MILES"].isin([23.1])] # rows where "MILES"==23.1
df[df["START"].isin(["Fort Pierce","Cary"])] # rows where "START" is Fort Pierce or Cary

df.loc[df["START"]=='Fort Pierce','START_DATE':'START'] # 4Columns rows where "START" is "Fort Pierce"
df.loc[df["START"].str.startswith("C")] # starting places name which starts with letter 'C'

incorrectIndex=df[(df["MILES"]>10)&(df["START"].isin(["Fort Pierce","Cary"]))].head(10)
```

# Proper Sequencing Index

```python
def sequenceIndex(incorrectIndex):
    incorrectIndex.reset_index(inplace=True,drop=True) # Index from 0 to last with step=1
    return incorrectIndex
sequenceIndex(incorrectIndex)
```

# Data Manipulation part-2

```python
sortedData1=df.sort_values(by=["MILES"],ascending=[False]) # Descending order sort
sortedData1.head()

 # 1st sort "START" in ascending order then sort "MILES" in descending order
sortedData2=df.sort_values(by=["START","MILES"],ascending=[True,False])
sequenceIndex(sortedData2.head(15)) # sortedData2 index won't be affected

# Creating a new column with 3 different categories("Short","Midium","long")
df["MILES_CAT"]=np.where(df["MILES"]<10,"Short",np.where(df["MILES"]<20,"Midium","long"))

# df["my_year"]=np.array(2020) #creates a new column "my_year" with value 2020 for all 

df.groupby("START")["MILES"].agg(["count","sum","mean","var","std"]) # aggregate values of "MILES" column
df.groupby("START").mean() # mean values of numerical data columns


df.groupby(["START","STOP"])["MILES"].agg(["mean","sum"]).head(10)
```

```python
rs=df.groupby("START").agg({"MILES":"mean","START_DATE":["min","max"]})
rs.reset_index(inplace=True)
rs.columns=["Start city","Avg dis","Most Past Trip","Recent Trip"]
rs.to_csv("fahim.csv")
```
